```python
#Import des librairies
import cv2
from ultralytics import YOLO

# Charger le modle Yolo dernire gnration
model = YOLO('yolov8n.pt')

# Ouvrir la vido 
video_path = "video/test_bmx_3.mp4"
cap = cv2.VideoCapture(video_path)

# Rcupration de la hauteur et largeur des images 
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
freq_ech = int(cap.get(cv2.CAP_PROP_FPS))

# Dfinition du codec de sortie de la vido
fourcc = cv2.VideoWriter_fourcc(*'XVID')

# Cration de la vido de sortie
output_video = cv2.VideoWriter('video/output_sprint.avi', fourcc, 30, (frame_width, frame_height))

# Loop lire les images unes par unes dans tant que la vido est ouverte
while cap.isOpened():
    # lecture du cadre vido 
    success, frame = cap.read()

    if success:
        # Application du modle YOLOv8 pour dtecter et suivre les objets
        results = model.track(frame, persist=True)

        # Rcupration de l'image annote
        annotated_frame = results[0].plot()

        # criture de l'image annote dans la vido de sortie
        output_video.write(annotated_frame)

        # Eventuellement afficher l'image annote
        # cv2.imshow("YOLOv8 Tracking", annotated_frame)

        # Sortir de la boucle si la touche "q" est presse
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Sortir de la boucle si la vido est termine
        break

# Fermer la vido et les fentres
cap.release()
output_video.release()
cv2.destroyAllWindows()
```
