```python
#Import des librairies
import cv2
from ultralytics import YOLO

# Charger le modèle Yolo dernière génération
model = YOLO('yolov8n.pt')

# Ouvrir la vidéo 
video_path = "video/test_bmx_3.mp4"
cap = cv2.VideoCapture(video_path)

# Récupération de la hauteur et largeur des images 
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
freq_ech = int(cap.get(cv2.CAP_PROP_FPS))

# Définition du codec de sortie de la vidéo
fourcc = cv2.VideoWriter_fourcc(*'XVID')

# Création de la vidéo de sortie
output_video = cv2.VideoWriter('video/output_sprint.avi', fourcc, 30, (frame_width, frame_height))

# Loop lire les images unes par unes dans tant que la vidéo est ouverte
while cap.isOpened():
    # lecture du cadre vidéo 
    success, frame = cap.read()

    if success:
        # Application du modèle YOLOv8 pour détecter et suivre les objets
        results = model.track(frame, persist=True)

        # Récupération de l'image annotée
        annotated_frame = results[0].plot()

        # Écriture de l'image annotée dans la vidéo de sortie
        output_video.write(annotated_frame)

        # Eventuellement afficher l'image annotée
        # cv2.imshow("YOLOv8 Tracking", annotated_frame)

        # Sortir de la boucle si la touche "q" est pressée
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Sortir de la boucle si la vidéo est terminée
        break

# Fermer la vidéo et les fenêtres
cap.release()
output_video.release()
cv2.destroyAllWindows()
```
